```python
# Check null value in dataset
data.isnull().sum()
sns.heatmap(data.isnull())
```

```python
# Perform data cleaning [Replace "?" with NaN]
data.head(20)
data.isin(["?"]).sum()
# data["workclass","occupation","native-country"]
data["workclass"] = data["workclass"].replace(to_replace="?", value=np.nan)
data["occupation"] = data["occupation"].replace(to_replace="?", value=np.nan)
data["native-country"] = data["native-country"].replace(to_replace="?", value=np.nan)
```

```python
# Let's check data now
data.isnull().sum()
sns.heatmap(data.isnull())
```

```python
# Drop all the missing value(check i percentage)
per_missing = data.isnull().sum()*100/len(data)
per_missing
data.dropna(how="any", inplace=True)   # any mean any missing value in row it will drop  ## inplace update the dataframe
data.shape
```

```python

# Check Duplicate data and drop it
dup = data.duplicated().any()
# print(f"Are there any duplicated values in data: {dup}")
data = data.drop_duplicates()
data.shape
```

```python
# Get Overall statistics (for all columns)
data.describe(include='all')
```

```python
# drop the columns ["educational-num","capital-gain","capital-loss"]
# print(data["education"].unique())
# print(data["educational-num"].unique())
data = data.drop(["educational-num","capital-gain","capital-loss"], axis=1)
data.columns
```
